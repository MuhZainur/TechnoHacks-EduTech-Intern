# Task 3 : Visualization using Histogram
## Create a histogram or bar chart to visualize the distribution of data in a dataset
### import dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_t3 = pd.read_csv(r"Iris_for_Task 3_5.csv")
df_t3.head()
```

### Check info, duplication, missing values and cleaning

```python
df_t3.info()
```

```python
# Duplicates values
df_t3.duplicated().sum()
```

```python
# Missing values
df_t3.isnull().sum()
```

No need cleaning, because data is already clean, let's make histogram for this data

```python
columns_hist = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
for column in columns_hist:
    df_t3[column].hist(color = 'purple', grid = False, bins = 20)
    plt.title(f"{column}")
    plt.show()
```

SepalLengthCm and SepalWidthCm almost have normal distribussion, but PetalLengthCm and PetalWidthCm have skewness.
